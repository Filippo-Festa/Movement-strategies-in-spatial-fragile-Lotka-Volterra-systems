```python
summation = 0

for i in tqdm(range(-N_p, N_p+1)):
    for j in range(-N_p, i+1):
        denom1 = lambd_0*(1 - (p_0 + (1-p_0)*((2*np.pi/N)*abs(i))**(beta-1))*(p_0 + (1-p_0)*((2*np.pi/N)*abs(j))**(beta-1)) ) + mu
        numer_x = np.sin((M+0.5)*l_0*(2*np.pi/N)*i)
        numer_y = np.sin((M+0.5)*l_0*(2*np.pi/N)*j)
        denom2_x = np.sin(l_0*(2*np.pi/N)*i)
        denom2_y = np.sin(l_0*(2*np.pi/N)*j)

        if i==0 and j==0:
            summation += (2*M + 1)**2/(denom1)
        elif i==0:
            summation += (numer_y*(2*M + 1))/(denom1*denom2_y)
        elif j==0:
            summation += (numer_x*(2*M + 1))/(denom1*denom2_x)
        else:
            summation += (numer_x*numer_y)/(denom1*denom2_x*denom2_y)

summation = summation*2/N**2
print(summation)
```

### Brute Force

```python
I = 0

for x in tqdm(range(-N_p, N_p+1)):
    for y in range(-N_p, N_p+1):

      if x==0 and y==0:
        I += (2*M+1)**2/mu


      elif x==0 and y!=0:
        f_y = f(2*np.pi*y/N, beta, tolerance)
        p_ky = (p0+(1-p0)*f_y)
        denom = lambd0*(1 -  p_ky ) + mu

        numer_y = np.sin( (2*M+1)*np.pi*l0*y/N )
        denom_y = np.sin( np.pi*l0*y/N )

        I += ((2*M+1)*numer_y)/(denom*denom_y)


      elif x!=0 and y==0:
        f_x = f(2*np.pi*x/N, beta, tolerance)
        p_kx = (p0+(1-p0)*f_x)
        denom = lambd0*(1 -  p_kx ) + mu

        numer_x = np.sin( (2*M+1)*np.pi*l0*x/N )
        denom_x = np.sin( np.pi*l0*x/N )

        I += ((2*M+1)*numer_x)/(denom*denom_x)


      else:
        f_x, f_y = f(2*np.pi*x/N, beta, tolerance), f(2*np.pi*y/N, beta, tolerance)
        p_kx, p_ky = (p0+(1-p0)*f_x),  (p0+(1-p0)*f_y)
        denom = lambd0*(1 -  p_kx*p_ky ) + mu

        numer_x = np.sin( (2*M+1)*np.pi*l0*x/N )
        numer_y = np.sin( (2*M+1)*np.pi*l0*y/N )
        denom_x = np.sin( np.pi*l0*x/N )
        denom_y = np.sin( np.pi*l0*y/N )

        I += (numer_x*numer_y)/(denom*denom_x*denom_y)


I = I/(N**2)                          # In ogni caso ci mette una vita a runnare, ci mette 3:30:29h... si può diminuire la discretizzazione nello spazio di fourier? Provare
I
```

### Even function

```python
I = 0

for x in tqdm(range(0, N_p+1)):
    for y in range(0, N_p+1):

      if x==0 and y==0:
        I += (2*M+1)**2/mu


      elif x==0 and y!=0:
        f_y = f(2*np.pi*y/N, beta, tolerance)
        p_ky = (p0+(1-p0)*f_y)
        denom = lambd0*(1 -  p_ky ) + mu

        numer_y = np.sin( (2*M+1)*np.pi*l0*y/N )
        denom_y = np.sin( np.pi*l0*y/N )

        I += 2*((2*M+1)*numer_y)/(denom*denom_y)


      elif x!=0 and y==0:
        f_x = f(2*np.pi*x/N, beta, tolerance)
        p_kx = (p0+(1-p0)*f_x)
        denom = lambd0*(1 -  p_kx ) + mu

        numer_x = np.sin( (2*M+1)*np.pi*l0*x/N )
        denom_x = np.sin( np.pi*l0*x/N )

        I += 2*((2*M+1)*numer_x)/(denom*denom_x)


      else:
        f_x, f_y = f(2*np.pi*x/N, beta, tolerance), f(2*np.pi*y/N, beta, tolerance)
        p_kx, p_ky = (p0+(1-p0)*f_x),  (p0+(1-p0)*f_y)
        denom = lambd0*(1 -  p_kx*p_ky ) + mu

        numer_x = np.sin( (2*M+1)*np.pi*l0*x/N )
        numer_y = np.sin( (2*M+1)*np.pi*l0*y/N )
        denom_x = np.sin( np.pi*l0*x/N )
        denom_y = np.sin( np.pi*l0*y/N )

        I += 4*(numer_x*numer_y)/(denom*denom_x*denom_y)


I = I/(N**2)
I
```
